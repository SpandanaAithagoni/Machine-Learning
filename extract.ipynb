This suggests that customers tend to tip more, on average, when riding in yellow taxis than in green taxis.

**HYPOTHESIS**
H0 is distribtution of tip for yellow = distribution of tip for green
H1 is distribution of tip for yellow != distribution of tip for green

```python
yellow_tip=data['tip'].loc[data['color']=='yellow']
green_tip=data['tip'][data['color']=='green']
```

```python
green_tip.shape
```

```python
yellow_tip.shape
```

```python
from scipy import stats
t_stat,p_val=stats.ttest_ind(yellow_tip,green_tip)
print("Independent T-Test statistic: ",t_stat)
print("P-value: ",p_val)

```

```python
alpha=0.05
if p_val<alpha:
    print("Reject null hypothesis")
else:
    print("Failed to reject the hypothesis")
```

**HYPOTHESIS**
H0 distribution of tip is the same across all pickup_borough categories 
H1 atleast one category has a different distribution of tip

```python
manhattan_tip=data['tip'][data['pickup_borough']=='Manhattan']
queens_tip=data['tip'][data['pickup_borough']=='Queens']
brooklyn_tip=data['tip'][data['pickup_borough']=='Brooklyn']
bronx_tip=data['tip'][data['pickup_borough']=='Bronx']

```

```python
f_stat,p_val=stats.f_oneway(manhattan_tip,queens_tip,brooklyn_tip,bronx_tip)
#print("ANOVA F-statistic: ",f_stat)
print("P-Value:",p_val)
alpha=0.05
if p_val<alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
    
    
```

```python
# ANOVA Test
regions = [group['tip'].values for name, group in data.groupby('pickup_zone')]
f_stat, p_val = stats.f_oneway(*regions)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")

```

```python
# ANOVA Test
regions = [group['tip'].values for name, group in data.groupby('dropoff_zone')]
f_stat, p_val = stats.f_oneway(*regions)
print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")

```

**INSIGHTS**
Pickup zone
If the p value is small then tips change across pickup zones.
If the p value is not small then tips are almost the same for all pickup zones.
Dropoff zone
If the p value is small then tips change across dropoff zones.
If the p value is not small then tips are almost the same for all dropoff zones.
